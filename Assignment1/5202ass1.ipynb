```python
def lower_clean_str(x):
  punc='!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
  lowercased_str = x.lower()
  for ch in punc:
    lowercased_str = lowercased_str.replace(ch, '')
  return lowercased_str

```

```python
l=sc.textFile('Scrum Handbook.txt')
one_RDD = l.map(lower_clean_str)
two_rdd = one_RDD.flatMap(lambda x: x.split(' ')).filter(lambda x: x!= ' ').filter(lambda x:x!= '').filter(lambda x:x!='•').map(lambda word: (word, 1))
two_rdd.collect()
```

```python
counts = two_rdd.reduceByKey(add)
```

```python
sorted_counts =  counts.sortBy(lambda x: x[1], False)
```

```python
i = 0

for word,count in sorted_counts.toLocalIterator():
        if(i<21):
            print(u"{} --> {}".format(word, count))
            i+=1
```

```python
stop_words = set(stopwords.words('english')) 
```

```python
stop_words_filtered = sorted_counts.filter(lambda word: word[0] not in stop_words)
```
