### Step 3

```python
def clean_str(x):
  symbols='!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
  string = x.lower()
  for char in symbols:
    string = string.replace(char, '')
  return string

```

```python
book1=sc.textFile('Scrum Handbook.txt')
#one_RDD = book1.map(lower_clean_str).flatMap(lambda x: x.split(' ')).filter(lambda x: x!= ' ').filter(lambda x:x!= '').filter(lambda x:x!='•')
book1_rdd = book1.map(clean_str).flatMap(lambda x: x.split(' ')).filter(lambda x: x!= ' ').filter(lambda x:x!= '').filter(lambda x:x!='•')

#one_RDD.collect()
book1_rdd.collect()
```

```python
book2=sc.textFile(r'Agile Processes  in Software Engineering  and Extreme Programming.txt')
book2_rdd = book2.map(clean_str).flatMap(lambda x: x.split(' ')).filter(lambda x: x!= ' ').filter(lambda x:x!= '').filter(lambda x:x!='•')

#one_RDD.collect()
book2_rdd.collect()
```

### Step 4

```python
# two_rdd = one_RDD.map(lambda word: (word, 1))
# two_rdd.collect()

book1_count_rdd = book1_rdd.map(lambda word: (word, 1))
book1_count_rdd.take(5)
```

```python
# two_rdd = one_RDD.map(lambda word: (word, 1))
# two_rdd.collect()

book2_count_rdd = book2_rdd.map(lambda word: (word, 1))
book2_count_rdd.take(5)
```

```python
counts_book1 = book1_count_rdd.reduceByKey(add)
```

```python
counts_book2 = book2_count_rdd.reduceByKey(add)
```

```python
sorted_counts_book1 =  counts_book1.sortBy(lambda x: x[1], False)
```

```python
sorted_counts_book2 =  counts_book2.sortBy(lambda x: x[1], False)
```

```python
i = 0

for word,count in sorted_counts_book1.toLocalIterator():
        if(i<21):
            print(u"{} --> {}".format(word, count))
            i+=1
```

```python
i = 0

for word,count in sorted_counts_book2.toLocalIterator():
        if(i<21):
            print(u"{} --> {}".format(word, count))
            i+=1
```

### Step 5

```python
stop_words = set(stopwords.words('english')) 
```

```python
stop_words_filtered_book1 = sorted_counts_book1.filter(lambda word: word[0] not in stop_words)
```

```python
stop_words_filtered_book1.take(10)
```

```python
stop_words_filtered_book2 = sorted_counts_book2.filter(lambda word: word[0] not in stop_words)
stop_words_filtered_book2.take(10)
```

```python
stop_words_filtered_book1.count()
```

```python
stop_words_filtered_book2.count()
```
