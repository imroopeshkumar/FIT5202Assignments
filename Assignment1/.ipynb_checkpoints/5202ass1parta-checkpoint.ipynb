## Step 2

```python

book1=sc.textFile(r'Scrum Handbook.txt') # read book1
book2=sc.textFile(r'Agile Processes  in Software Engineering  and Extreme Programming.txt') # read book2
```

### Step 3

```python
def clean_str(x): # function to clean punctuations
  symbols='!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~0123456789■'
  string = x.lower()
  for char in symbols:
    string = string.replace(char, '')
  return string

```

```python
#one_RDD = book1.map(lower_clean_str).flatMap(lambda x: x.split(' ')).filter(lambda x: x!= ' ').filter(lambda x:x!= '').filter(lambda x:x!='•')
book1_rdd = book1.map(clean_str).flatMap(lambda x: x.split(' ')).filter(lambda x: x!= ' ').filter(lambda x:x!= '').filter(lambda x:x!='•') # clean book1

#one_RDD.collect()
book1_rdd.collect()
```

```python
book2_rdd = book2.map(clean_str).flatMap(lambda x: x.split(' ')).filter(lambda x: x!= ' ').filter(lambda x:x!= '').filter(lambda x:x!='•') #clean book 2

#one_RDD.collect()
book2_rdd.collect()
```

### Step 4

```python
# two_rdd = one_RDD.map(lambda word: (word, 1))
# two_rdd.collect()

book1_count_rdd = book1_rdd.map(lambda word: (word, 1)) # map each word with number 1
book1_count_rdd.take(5)
```

```python
# two_rdd = one_RDD.map(lambda word: (word, 1))
# two_rdd.collect()

book2_count_rdd = book2_rdd.map(lambda word: (word, 1)) # map each word with number 1
book2_count_rdd.take(5)
```

```python
counts_book1 = book1_count_rdd.reduceByKey(add)
```

```python
counts_book2 = book2_count_rdd.reduceByKey(add)
```

```python
sorted_counts_book1 =  counts_book1.sortBy(lambda x: x[1], False) # count and sort 
```

```python
sorted_counts_book2 =  counts_book2.sortBy(lambda x: x[1], False) # count and sort
```

```python
i = 0

for word,count in sorted_counts_book1.toLocalIterator(): # print top 21 words book1
        if(i<21):
            print(u"{} --> {}".format(word, count))
            i+=1
```

```python
i = 0

for word,count in sorted_counts_book2.toLocalIterator(): # print top 21 words book2
        if(i<21):
            print(u"{} --> {}".format(word, count))
            i+=1
```

### Step 5

```python
stop_words = set(stopwords.words('english')) # english stop words
```

```python
stop_words_filtered_book1 = sorted_counts_book1.filter(lambda word: word[0] not in stop_words) # remove stop words book1
```

```python
stop_words_filtered_book1.take(10)
```

```python
stop_words_filtered_book2 = sorted_counts_book2.filter(lambda word: word[0] not in stop_words) # remove stop words book2
stop_words_filtered_book2.take(10)
```

```python
stop_words_filtered_book1.count() # number of words book1
```

```python
stop_words_filtered_book2.count() # number of words book2
```

### Step 6

```python
i = 0
wordcount = 0
for word,count in stop_words_filtered_book1.toLocalIterator():
#             print(u"{} --> {}".format(word, count))
            wordcount+=count
    
print(wordcount)
```

```python
avgwordoccurance_book1 = wordcount/stop_words_filtered_book1.count()
avgwordoccurance_book1 # avg word occurance book1
```

```python
i = 0
wordcount = 0
for word,count in stop_words_filtered_book2.toLocalIterator():
#             print(u"{} --> {}".format(word, count))
            wordcount+=count
    
print(wordcount)
```

```python
avgwordoccurance_book2 = wordcount/stop_words_filtered_book2.count()
avgwordoccurance_book2 # avg word occurance book2
```

```python
top15_stop_words_filtered_book1_array = []
top15_stop_words_filtered_book1_arraycount = []
top15array = stop_words_filtered_book1.take(15)
type(top15array)
for x in top15array:
#     print(x[1])
    
    top15_stop_words_filtered_book1_array.append(x[0])
    top15_stop_words_filtered_book1_arraycount.append(x[1])
```
