```python
# import statements
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt
import csv


def read_csv(fileName):
    data = []
    cRows = csv.DictReader(open(fileName))
    for rows in cRows:
        data.append(rows)
    return data

def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, data)
        print('Message published successfully. Data: ' + str(data))
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x: dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    topic = 'memory'
    process_rows = read_csv('../../Streaming_Linux_memory.csv')
    
    print('Publishing records..')
    producer = connect_kafka_producer()
    
    for i in range(len(process_rows)):
        if process_rows[i]["machine"] == "4":
            row_start_4 = i
            row_end_4 = row_start_4
            break
            
    for i in range(len(process_rows)):
        if process_rows[i]["machine"] == "5":
            row_start_5 = i
            row_end_5 = row_start_5
            break
    
    for i in range(len(process_rows)):
        if process_rows[i]["machine"] == "6":
            row_start_6 = i
            row_end_6 = row_start_6
            break
            
    for i in range(len(process_rows)):
        if process_rows[i]["machine"] == "7":
            row_start_7 = i
            row_end_7 = row_start_7
            break
    
    for i in range(len(process_rows)):
        if process_rows[i]["machine"] == "8":
            row_start_8 = i
            row_end_8 = row_start_8
            break
    
    data = []
    
    while(True):
        row_count_x_4 = random.randint(20,80)
        data_4 = process_rows[row_end_4:row_end_4+row_count_x_4]
        row_end_4 = row_end_4 + row_count_x_4  
        
        row_count_x_5 = random.randint(20,80)
        data_5 = process_rows[row_end_5:row_end_5+row_count_x_5]
        row_end_5 = row_end_5 + row_count_x_5
        
        row_count_x_6 = random.randint(20,80)
        data_6 = process_rows[row_end_6:row_end_6+row_count_x_6]
        row_end_6 = row_end_6 + row_count_x_6
        
        row_count_x_7 = random.randint(20,80)
        data_7 = process_rows[row_end_7:row_end_7+row_count_x_7]
        row_end_7 = row_end_7 + row_count_x_7
        
        row_count_x_8 = random.randint(20,80)
        data_8 = process_rows[row_end_8:row_end_8+row_count_x_8]  
        row_end_8 = row_end_8 + row_count_x_8
        
        for i in data_4:
            i["ts"] = int(dt.datetime.utcnow().timestamp())
            data.append(i)  
        for i in data_5:
            i["ts"] = int(dt.datetime.utcnow().timestamp())
            data.append(i)
        for i in data_6:
            i["ts"] = int(dt.datetime.utcnow().timestamp())
            data.append(i)
        for i in data_7:
            i["ts"] = int(dt.datetime.utcnow().timestamp())
            data.append(i)
        for i in data_8:
            i["ts"] = int(dt.datetime.utcnow().timestamp())
            data.append(i)
                    
        publish_message(producer, topic, data)
        
        data = []
        
        row_count_y_4 = random.randint(0,5)
        data_4 = process_rows[row_end_4:row_end_4+row_count_y_4]
        row_end_4 = row_end_4 + row_count_y_4

        row_count_y_5 = random.randint(0,5)
        data_5 = process_rows[row_end_5:row_end_5+row_count_y_5]
        row_end_5 = row_end_5 + row_count_y_5
        
        row_count_y_6 = random.randint(0,5)
        data_6 = process_rows[row_end_6:row_end_6+row_count_y_6]
        row_end_6 = row_end_6 + row_count_y_6
        
        row_count_y_7 = random.randint(0,5)
        data_7 = process_rows[row_end_7:row_end_7+row_count_y_7]
        row_end_7 = row_end_7 + row_count_y_7
        
        row_count_y_8 = random.randint(0,5)
        data_8 = process_rows[row_end_8:row_end_8+row_count_y_8]
        row_end_8 = row_end_8 + row_count_y_8
        
        for i in data_4:
            i["ts"] = int(dt.datetime.utcnow().timestamp())
            data.append(i)  
        for i in data_5:
            i["ts"] = int(dt.datetime.utcnow().timestamp())
            data.append(i)
        for i in data_6:
            i["ts"] = int(dt.datetime.utcnow().timestamp())
            data.append(i)
        for i in data_7:
            i["ts"] = int(dt.datetime.utcnow().timestamp())
            data.append(i)
        for i in data_8:
            i["ts"] = int(dt.datetime.utcnow().timestamp())
            data.append(i)        
        
        if(row_end_4+row_end_5+row_end_6+row_end_7+row_end_8 >= len(process_rows)):
            row_end_4 = row_start_4
            row_end_5 = row_start_5
            row_end_6 = row_start_6
            row_end_7 = row_start_7
            row_end_8 = row_start_8
        sleep(10)
        print('------------------------------------------------------------')
        
```
