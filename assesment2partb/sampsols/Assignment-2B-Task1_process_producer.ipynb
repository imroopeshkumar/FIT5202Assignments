```python
# import statements
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt
import csv


def read_csv(fileName):
    data = []
    
    cRows = csv.DictReader(open(fileName))
    for rows in cRows:
        data.append(rows)
    return data

def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, data)
        print('Message published successfully. Data: ' + str(data))
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x: dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    topic = 'process'
    process_rows = read_csv('../Streaming_Linux_process.csv')
    
    print('Publishing records..')
    producer = connect_kafka_producer()
    
    for i in range(len(process_rows)):
        if process_rows[i]["machine"] == "4":
            row_start_4 = i
            row_end_4 = row_start_4
            break
            
    for i in range(len(process_rows)):
        if process_rows[i]["machine"] == "5":
            row_start_5 = i
            row_end_5 = row_start_5
            break
    
    for i in range(len(process_rows)):
        if process_rows[i]["machine"] == "6":
            row_start_6 = i
            row_end_6 = row_start_6
            break
            
    for i in range(len(process_rows)):
        if process_rows[i]["machine"] == "7":
            row_start_7 = i
            row_end_7 = row_start_7
            break
    
    for i in range(len(process_rows)):
        if process_rows[i]["machine"] == "8":
            row_start_8 = i
            row_end_8 = row_start_8
            break
    count = 1
    while(True):
        data=[]
        row_count_4 = random.randint(10,50)
        data4 = process_rows[row_end_4:row_end_4+row_count_4]
        row_end_4 = row_end_4 + row_count_4
        row_count_5 = random.randint(10,50)
        data5 = process_rows[row_end_5:row_end_5+row_count_5]
        row_end_5 = row_end_5 + row_count_5
        row_count_6 = random.randint(10,50)
        data6 = process_rows[row_end_6:row_end_6+row_count_6]
        row_end_6 = row_end_6 + row_count_6
        row_count_7 = random.randint(10,50)
        data7 = process_rows[row_end_7:row_end_7+row_count_7]
        row_end_7 = row_end_7 + row_count_7
        row_count_8 = random.randint(10,50)
        data8 = process_rows[row_end_8:row_end_8+row_count_8]  
        row_end_8 = row_end_8 + row_count_8
        
        for i in data4:
            data.append(i)  
        for i in data5:
            data.append(i)
        for i in data6:
            data.append(i)
        for i in data7:
            data.append(i)
        for i in data8:
            data.append(i)
            
        ts = int(dt.datetime.utcnow().timestamp())    
        data_full = {'data':data, 'ts':ts,'count':count}
        publish_message(producer, topic, data_full)
        count +=1
        if(row_end_4+row_end_5+row_end_6+row_end_7+row_end_8 >= len(process_rows)):
            row_end_4 = row_start_4
            row_end_5 = row_start_5
            row_end_6 = row_start_6
            row_end_7 = row_start_7
            row_end_8 = row_start_8
            count = 0
        
        sleep(5)
        
```
