```python
# import statements
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt
import csv


def read_csv(fileName): #Function to read a csv file as a dictionary with filename as the input parameter
    data = []
    
    cRows = csv.DictReader(open(fileName))
    for rows in cRows:
        data.append(rows)
    return data

def publish_message(producer_instance, topic_name, data): #Function to publish the message
    try:
        producer_instance.send(topic_name, data)
        print('Message published successfully')
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x: dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    topic = 'process' #Defining the topic name
    process_rows = read_csv('../Streaming_Linux_process.csv') #reading the csv and getting a dictionary using a defined function
    
    print('Publishing records..')
    
    producer = connect_kafka_producer()
    
    data4 = [] #Initialising an empty list to store the data of machine 4
    data5 = [] #Initialising an empty list to store the data of machine 5
    data6 = [] #Initialising an empty list to store the data of machine 6
    data7 = [] #Initialising an empty list to store the data of machine 7
    data8 = [] #Initialising an empty list to store the data of machine 8
    
    for i in range(len(process_rows)):  #Changing the datatypes of the elements to correct datatype
        process_rows[i]["sequence"] = int(process_rows[i]["sequence"])
        process_rows[i]["machine"] = int(process_rows[i]["machine"])
        process_rows[i]["PID"] = int(process_rows[i]["PID"])
        process_rows[i]["TRUN"] = int(process_rows[i]["TRUN"])
        process_rows[i]["TSLPI"] = int(process_rows[i]["TSLPI"])
        process_rows[i]["TSLPU"] = int(process_rows[i]["TSLPU"])
        process_rows[i]["POLI"] = str(process_rows[i]["POLI"])
        process_rows[i]["NICE"] = int(process_rows[i]["NICE"])
        process_rows[i]["PRI"] = int(process_rows[i]["PRI"])
        process_rows[i]["RTPR"] = int(process_rows[i]["RTPR"])
        process_rows[i]["CPUNR"] = int(process_rows[i]["CPUNR"])
        process_rows[i]["EXC"] = int(process_rows[i]["EXC"])
        process_rows[i]["Status"] = str(process_rows[i]["Status"])
        process_rows[i]["State"] = str(process_rows[i]["State"])
        process_rows[i]["CPU"] = float(process_rows[i]["CPU"])
        process_rows[i]["CMD"] = str(process_rows[i]["CMD"])
        
    for i in range(len(process_rows)): #getting the data for each machine and appending to the respective lists
        if process_rows[i]["machine"] == 4:
            data4.append(process_rows[i])
        if process_rows[i]["machine"] == 5:
            data5.append(process_rows[i])
        if process_rows[i]["machine"] == 6:
            data6.append(process_rows[i])
        if process_rows[i]["machine"] == 7:
            data7.append(process_rows[i])
        if process_rows[i]["machine"] == 8:
            data8.append(process_rows[i])
    #set all index 0        
    rowstart4 = 0 
    rowstart5 = 0
    rowstart6 = 0
    rowstart7 = 0
    rowstart8 = 0   
    
    while(True):
        data=[] #init empty list to store the final data from single message
        
        rowcount4 = random.randint(10,50) #get random  machine 4
        rowend4 = rowstart4 + rowcount4 
        if rowend4 > len(data4)-1: 
            newdata4 = data4[rowstart4:len(data4)-1] 
            rowend4 = rowcount4 - len(newdata4)-1 
            rowstart4 = 0 
            newcountdata4 = data4[rowstart4:rowend4] 
            for i in newcountdata4:
                newdata4.append(i)
        elif rowend4 == len(data4)-1: #if reached last item
            newdata4 = data4[rowstart4:rowend4]
            rowend4 = -1 
        else:
            newdata4 = data4[rowstart4:rowend4]
        rowstart4 = rowend4 + 1 
        
        rowcount5 = random.randint(10,50) #do same for machine = 5
        rowend5 = rowstart5 + rowcount5 #Getting the end index
        if rowend5 > len(data5)-1: 
            newdata5 = data5[rowstart5:len(data5)-1] 
            rowend5 = rowcount5 - len(newdata5)-1 
            rowstart5 = 0 
            newcountdata5 = data5[rowstart5:rowend5]
            for i in new_countdata5:
                newdata5.append(i)
        elif rowend5 == len(data5)-1:
            newdata5 = data5[rowstart5:rowend5]
            rowend5 = -1 
        else:
            newdata5 = data5[rowstart5:rowend5]
        rowstart5 = rowend5 + 1 
        
        rowcount6 = random.randint(10,50) # do same for machine 6
        rowend6 = rowstart6 + rowcount6 #Getting the end index
        if rowend6 > len(data6)-1: 
            newdata6 = data6[rowstart6:len(data6)-1] 
            rowend6 = rowcount6 - len(newdata6)-1 
            rowstart6 = 0 
            newcountdata6 = data4[rowstart6:rowend6]
            for i in newcountdata6:
                newdata6.append(i)
        elif rowend6 == len(data6)-1: 
            newdata6 = data6[rowstart6:rowend6]
            rowend6 = -1 
        else:
            newdata6 = data6[rowstart6:rowend6]
        rowstart6 = rowend6 + 1 
        
        rowcount7 = random.randint(10,50) #do same for machine = 7
        rowend7 = rowstart7 + rowcount7 
        if rowend7 > len(data7)-1: 
            newdata7 = data7[rowstart7:len(data7)-1] 
            rowend7 = rowcount7 - len(newdata7)-1 
            rowstart7 = 0 
            newcountdata7 = data7[rowstart7:rowend7]
            for i in newcountdata7:
                newdata7.append(i)
        elif rowend7 == len(data7)-1: 
            newdata7 = data7[rowstart7:rowend7]
            rowend7 = -1 
        else:
            newdata7 = data7[rowstart7:rowend7]
        rowstart7 = rowend7 + 1 #Reset the  next index of last index as the next start index
        
        rowcount8 = random.randint(10,50) #do same for machine 8 
        rowend8 = rowstart8 + rowcount8 
        if rowend8 > len(data8)-1: 
            newdata8 = data8[rowstart8:len(data8)-1] 
            rowend8 = rowcount8 - len(newdata8)-1 
            rowstart_8 = 0 
            newcountdata8 = data8[rowstart8:rowend8]
            for i in newcountdata8:
                newdata8.append(i)
        elif rowend8 == len(data8)-1: 
            newdata8 = data8[rowstart8:rowend8]
            rowend8 = -1 
        else:
            newdata8 = data8[rowstart8:rowend8]
        rowstart8 = rowend8 + 1 
        
        ts = int(dt.datetime.utcnow().timestamp())
        
        #append ts
        for i in newdata4:
            i["ts"] = ts
            data.append(i)  
        for i in newdata5:
            i["ts"] = ts
            data.append(i)
        for i in newdata6:
            i["ts"] = ts
            data.append(i)
        for i in newdata7:
            i["ts"] = ts
            data.append(i)
        for i in newdata8:
            i["ts"] = ts
            data.append(i)
        
        publish_message(producer, topic, data) #Publish the data
        
        sleep(5) #Wait for 5 seconds to start next cycle
            
        
```
